```python
#Online Book Store using obkect oriented programming
class Books:
    def __init__(self, title, price, stock):
        self.title = title
        self.price = price
        self.stock = stock
    def display_info(self):
        return f"{self.title} --> its price ${self.price}"
        
class Customers:
    def __init__(self,name):
        self.name = name
        self.cart = []
    def add_to_cart(self,book):
        self.cart.append(book)
        print(f"{book.title} is added to your cart..\n")
    def view_cart(self):
        if not self.cart:
            print("Your cart is empty...")
        else:
            print("Your cart:")
        for book in self.cart:
            print(f"{book.display_info()}")

class Order:
    def __init__(self, customer, books):
        self.customer = customer
        self.books = books
        
    def complete_order(self):
        total_price = sum(book.price for book in self.books)
        if total_price == 0:
            print("Your cart is empty.Please add bookks to yur cart...\n")
        elif total_price > 0:
            print(f"Order completed for {self.customer.name}. Total Amount: ${total_price}\n")
            for book in self.books:
                book.stock -= 1
            self.customer.cart = []
        else:
            print("Inalid total amount.Please check your cart...")

b1 = Books("Python Programming", 20,4)
b2 = Books("Probability and statistics", 19.5,5)
b3 = Books("Machine Learning", 25,6)
b4 = Books("Power BI", 23.5,2)
b5 = Books("Big data analysis",23.5,4)
b6 = Books("Deep Learning", 24.5,5)
c = Customers("Ajinkya")

while True:
    print("\nWelcome to the Online Bookstore!")
    print("1. Browse Books")
    print("2. View Cart")
    print("3. Complete Order")
    print("4. Exit")
    
    choice = input("Enter Your Choice(1-4):")
    
    if choice == "1":
        books = [b1, b2, b3, b4, b5, b6]
        print("Available Books are:")
        for idx, book in enumerate(books, start=1):
            print(f"{idx}. {book.display_info()}")
        book_choice = int(input("Enter the number of the book you want to add to your cart (0 to o back)"))
        if 1 <= book_choice <= len(books):
            c.add_to_cart(books[book_choice - 1])
        elif book_choice == 0:
            continue
        else:
            print("Invalid choice. Please try again.")
            
    elif choice == "2":
        c.view_cart()
    
    elif choice == "3":
        order = Order(c,c.cart)
        order.complete_order()
    
    elif choice == "4":
        print("Thank you for shopping with us. Goodbye!")
        break
    
    else:
        print("Invalid choice. Please try again... ")
        
```
